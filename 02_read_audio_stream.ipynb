```python
import soundcard as sc
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
# get a list of all speakers:
speakers = sc.all_speakers()
# get the current default speaker on your system:
default_speaker = sc.default_speaker()
# get a list of all microphones:
mics = sc.all_microphones()
# get the current default microphone on your system:
default_mic = sc.default_microphone()
```

```python
data = default_mic.record(samplerate=44100, numframes=44100)

%matplotlib inline
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(np.linspace(0, 1, num=44100), data)

# Attenuate the initial noise
data[:int(44100*0.05)] = 0
ax2.plot(np.linspace(0, 1, num=44100), data)
```

## Read audio stream continuously

```python
with default_mic.recorder(samplerate=48000) as mic:
    for _ in range(100):
        data = mic.record(numframes=1024)
        sp.play(data)
```

```python
with mics[1].recorder(samplerate=48000) as mic, \
      default_speaker.player(samplerate=48000) as sp:
    for _ in range(500):
        data = mic.record(numframes=1024)
        sp.play(data)
```
